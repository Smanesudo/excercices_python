# substrings

Output all the possible substrings of a given string.

```python
#version list
def substring_list(s):
    l = []
    n = len(s)
    for i in range(1, n+1):
        for j in range(n-i+1):
            l.append(s[j:j+i])
    return l

#version set
def substring_set(s):
    l = set()
    n = len(s)
    for i in range(1, n+1):
        for j in range(n-i+1)
            l.add(s[j:j+i])
    return l

#version list comprehension
def substring_comprehension(s):
    return [s[j:j+i] for i in range(1, len(s)+1)
            for j in range(len(s)-i+1)]
```

```python
substring_set('manger')
```

```python
substring_list('manger')
```

```python
substring_comprehension('manger')
```

# fibonacci

```python
def fibonacci(n):
    #     if n==0 or n==1:
    #         return 1
    a, b = 1, 1
    for i in range(n):
        a, b = b, a+b
    return a
```

```python
for i in range(100):
    print(fibonacci(i))
```

# Kth Largest Element in Array

```python
def klargest(l, k):
    return sorted(l)[k-1]


def kth_largest(l, k):
    l.sort()
    return l[k-1]
```

# factorial

```python

# naive

def factorial(n):
    if n == 0:
        return 1
    res = 1
    for i in range(1, n+1):
        res *= i
    return res

# recurrence
def fact_rec(n):
    if n == 0 or n == 1:
        return 1
    return n*fact_rec(n-1)
```

```python
factorial(5000)
```

```python
fact_rec(5000)
```

# Largest Continuous 

Given an array of integers (positive and negative) find the largest continuous sum.

```python
def largestContinuousSum(arr):
    if len(arr) == 0:
        return
    maxSum = currentSum = arr[0]
    for num in arr[1:]:
        currentSum = max(currentSum+num, num)
        maxSum = max(currentSum, maxSum)
    return maxSum
```

```python
largestContinuousSum([-10, 2, 3, -1, 4, -2, 3, -1])
```

# room number 

You're in a hotel and you forgot what room number you were in but remember that the sum of it's divisors is greater than the number, yet there is no subset of those divisors that add up to the number itself. There are 100 rooms in the Hotel, what's your room number?

```python
from math import sqrt


def divisors(n):
    return [i for i in range(1, int(n/2)+1) if n % i == 0]


def subsets(l):
    for start in range(len(l)-1):
        for end in range(start+1, len(l)-1):
            yield l[start:end+1]


def get_room_nr(rooms):
    for roomnr in range(1, rooms+1):
        #         print(roomnr)
        div = divisors(roomnr)
        if sum(div) <= roomnr:
            continue
#         print(roomnr)
        found = False
        for s in subsets(div):
            #             print(s)
            if sum(s) == roomnr:
                found = True
#                 print(s)
                break
        if not found:
            return roomnr


print(get_room_nr(100))
```

```python
def ispalindrome_shorter(s):

    return True if s == s[::-1] else False


def largest_palindrome(s):
    n = len(s)
    if n <= 1:
        return s
    for size_sub in range(n, 0, -1):
        for i in range(n-size_sub+1):
            if ispalindrome_shorter(s[i:i+size_sub]):
                return s[i:i+size_sub]
```

```python
largest_palindrome('I am a red racecar driver')
```
